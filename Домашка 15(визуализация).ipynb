```python
import pandas as pd
import matplotlib.pyplot as plt
```

# База данных 'Netflix TV Shows and Movies'

https://www.kaggle.com/datasets/victorsoeiro/netflix-tv-shows-and-movies

```python
df = pd.read_csv('/Users/anastasiafedoracenko/Desktop/titles.csv')
df.head(60)
```

```python
df.info()
```

```python
df1 = df[['age_certification', 'title']].groupby('age_certification').count()
df1
```

```python
df1.plot( kind = 'bar', title = 'Количество фильмов и сериалов нетфликс по возрастным ограничениям')


# Можно отследить какое количество фильмов было снять в определенное возрастной категории за все время 
```

```python
df2 = df[['release_year', 'title']].groupby('release_year').count()
df2
```

```python
plt.plot(df2.index, df2['title'])
plt.title('Динамика релизов на Netflix')
plt.xlabel('Год')
plt.ylabel('Количество релизов')

#Можно отследить, что до 2000х годов количество релизов не превышало ста, в 2020 году было рекордное количество
# релизов, на данный момент 2022 не закончен, полагаю, что в этом году колечество релизов вырастет 
```

```python
df3 = df[['type', 'title']].groupby('type').count()
df3
```

```python
df3.plot( kind = 'bar', title = 'Количество фильмов и сериалов нетфликс за все время')

# за все время было выпущено в два раза больше фильмов, чем сериалов
```

```python
df4 = df[['title', 'imdb_score', 'tmdb_score']].groupby('title').sum()
df4.head()
```

```python
df4.loc[['#realityhigh', '3 Ninjas Kick Back', '#AnneFrank. Parallel Stories', '#Alive', '#FriendButMarried 2', '13 Reasons Why']].plot( kind = 'bar', title = 'Соотношение оценок на 6 рандомных релизах от нетфликс')

# можно увидеть, что оценки на TMDB выше, чем на IMDB
```
